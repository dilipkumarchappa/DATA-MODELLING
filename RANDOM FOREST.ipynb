```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def myvif(x):
    data= x.as_matrix()
    vif = [variance_inflation_factor(data, i) for i in range(data.shape[1])]
    vif_factors=pd.DataFrame()
    vif_factors['columns']=x.columns
    vif_factors['vif']=vif
    return vif_factors
```
